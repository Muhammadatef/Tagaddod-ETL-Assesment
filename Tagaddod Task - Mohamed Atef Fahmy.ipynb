## Step two IN ETL : Transform
###### a function to implement and convert latitude & longitude to a location or address so can analytics team make the required map

```python
def transform():
    latitude = list[df2['latitude']]
    longitude = list[df2['longitude']]


    for latitude, longitude in df2.itertuples(index=False):

        # calling the nominatim tool
        geoLoc = Nominatim(user_agent="GetLoc")

        # passing the coordinates
        locname = geoLoc.reverse([latitude,longitude])

        # printing the address/location name
        Location = locname.address
        print(locname.address)
    return Location
```
